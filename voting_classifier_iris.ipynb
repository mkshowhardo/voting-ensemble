```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
cl1 = LogisticRegression()
cl2 = KNeighborsClassifier()
cl3 = RandomForestClassifier()
```

```python
estimators = [('lr1',cl1),('kn2',cl2),('rf3',cl3)]
```

```python
#Calculate how many samples to remove
num_samples_to_remove = x.shape[0] % 10

# Undersample by removing rows from x and y
x = new_df.iloc[:,0:2]  # Remove rows from the end
y = new_df.iloc[:,-1]  # Remove rows from the end

# Continue with your loop
for estimator in estimators:
    scores = cross_val_score(estimator[1], x, y, cv=10, scoring='accuracy')
    print(estimator[0], np.round(np.mean(scores), 2))
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
vc = VotingClassifier(estimators = estimators,voting='hard')
x = cross_val_score(vc,x,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))
```

```python
from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
import numpy as np

x, y = make_classification(n_samples=1000, n_features=20, n_informative=15, n_redundant=5, random_state=2)

svm1 = SVC(kernel='poly', degree=1, probability=True)
svm2 = SVC(kernel='poly', degree=2, probability=True)
svm3 = SVC(kernel='poly', degree=3, probability=True)
svm4 = SVC(kernel='poly', degree=4, probability=True)
svm5 = SVC(kernel='poly', degree=5, probability=True)
svm6 = SVC(kernel='poly', degree=6, probability=True)

estimators = [('svm1', svm1), ('svm2', svm2), ('svm3', svm3), ('svm4', svm4), ('svm5', svm5), ('svm6', svm6)]

for estimator in estimators:
    scores = cross_val_score(estimator[1], x, y, cv=10, scoring='accuracy')  # Change variable name to 'scores'
    print(estimator[0], np.round(np.mean(scores), 2))
```

```python

vc1 = VotingClassifier(estimators=estimators, voting='soft')
scores = cross_val_score(vc1, x, y, cv=10, scoring='accuracy')  # Use lowercase 'x' and assign to 'scores'
print(np.round(np.mean(scores), 2))
```
